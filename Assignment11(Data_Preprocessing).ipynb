Importing the libraries

```python
import numpy as np
import pandas as pd
```

Importing dataset

```python
data=pd.read_csv("Data.csv")
data
```

Handling the missing data

```python
data.isnull().sum()
```

```python
#there are nan values in age and salary columns
#we replace the nan values with median
data["Age"].fillna(data["Age"].median(),inplace=True)
data["Salary"].fillna(data["Salary"].median(),inplace=True)
data.isnull().sum()
```

```python
data
```

Encoding categorical data



```python
#Purchased column has binary categorical values so we do label encoding
from sklearn import preprocessing
data["Purchased"]=data[["Purchased"]].apply(preprocessing.LabelEncoder().fit_transform)
data
```

 Creating a dummy variable

```python
df_onehot=pd.get_dummies(data)
df_onehot
```

Splitting the dataset into training sets and Test sets

```python
from sklearn.model_selection import train_test_split

X=df_onehot.drop("Purchased",axis=1)
y=df_onehot[["Purchased"]]

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
X_train.shape
```

```python
y_train.shape
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()#initialize
# we are finding the values of mean and sd and transform the training data
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled
```
